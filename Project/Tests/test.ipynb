```python
import string

content = []
with open('dracula.txt', 'r') as f:
    index = 0
    for line in f.readlines():
        content.append(line)

catalogue_start = content.index('CONTENTS\n')
catalogue_end = content.index('DRACULA\n')
print(catalogue_start, catalogue_end)

catalogue = content[catalogue_start: catalogue_end]
chapters = []
for line in catalogue:
    if line.startswith('CHAPTER'):
        chapters.append(line.strip())


titles = []
for line in catalogue:
    if not line.startswith('CHAPTER') and not line.startswith('CONTENTS\n') and not line == '\n' and "Page" not in line:
        title = line.strip()
        for punc in string.punctuation:
            title = title.replace(punc, '')
        title = '_'.join(title.split()[:-1])
        titles.append(title)

chapters_index = []
for chapter in chapters:
    chapters_index.append(content[catalogue_end:].index(chapter+'\n'))
chapters_index.append(-1)

chapter_content = []
offset = catalogue_end
for index in range(1, len(chapters_index)):
    chapter_start, chapter_end = chapters_index[index-1]+offset, chapters_index[index]+offset
    chapter_content.append(content[chapter_start: chapter_end])

len(chapter_content)

counter = 1
for chapter in chapter_content:
    with open("chapters/Dracula-Chapter-{}-{}.txt".format(counter, titles[counter-1]), 'w', encoding='utf-8') as f:
        f.writelines(chapter)
    counter += 1




```
